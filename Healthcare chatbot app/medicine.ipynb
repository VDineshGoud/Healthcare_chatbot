## Disease symptoms DataSet

```python
df3.head()
```

```python
df3.Disease.nunique()
```

```python
data_01 = df3.fillna("@")
data_01.head()
```

```python
def preprocess(raw_text):
    # Removing special characters and digits
    sentence = re.sub("[^a-zA-Z]", " ", raw_text)
    
    # change sentence to lower case
    sentence = sentence.lower()

    # tokenize into words
    tokens = sentence.split()
    
    return " ".join(tokens)

for i in data_01.columns:
    data_01[i] = data_01[i].apply(lambda x : preprocess(x))

data_01.head()
```

```python
for i in data_01.columns:
    data_01[i].replace('', 0, inplace = True)
    
data_01.head()
```

```python
disease = data_01['Disease'].unique()
symptoms = []
for i in data_01.columns:
    symptoms.extend(data_01[i].unique())
    
        
symptoms = [x for x in symptoms if x!=0 ]

# symptoms
```

```python
dict_01 = {}
for i in disease:
    sym = []
    temp = data_01[data_01['Disease']== i ].iloc[0:, 1:].values
    for j in temp:
        t1 = []
        for k in j:
            if (k != 0):
                t1.append(k)
        sym.extend(t1)
        
    dict_01[i] = sym
    

dict_01
```

```python
json_object = json.dumps(dict_01, indent = 4) 
print(json_object)
```

```python
with open('Disease_symptoms.json', 'w') as fp:
    json.dump(dict_01, fp)
```
